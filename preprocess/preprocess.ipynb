```python
import os
import random
batch_1_path = './dataset/part_1/'
batch_1_list = [batch_1_path+i+'/' for i in os.listdir(batch_1_path)]
batch_2_path = './dataset/part_2/'
batch_2_list = [batch_2_path+i+'/' for i in os.listdir(batch_2_path)]
```

```python
def process_data(batch_list):
    abstract = []
    introduction = []
    conclusion = []
    summary = []
    for path in batch_list:
        if 'DS_Store' in path:
            continue
            
        # 处理abstract
        abstract_path = path+'ABSTRACT'
        data = open(abstract_path)
        lines = data.readlines()
        lines = [i.strip('\n') for i in lines]
        lines = [i for i in lines if i != '']
        try:
            lines = lines[lines.index('PARAGRAPH')+1:]
            lines = [line for line in lines if line!='PARAGRAPH']
            one_abstract = " ".join(lines)
        except:
            one_abstract = ""
            print(abstract_path, 'no abstract')
        abstract.append(one_abstract)
        
        # 处理introduction
        full_path = path+'FULLTEXT'
        data = open(full_path)
        lines = data.readlines()
        lines = [i.strip('\n') for i in lines]
        lines = [i for i in lines if i != '']
        try: 
            lines = lines[lines.index('Introduction')+2:]
            one_intro_list = []
            for line in lines:
                if line == 'PARAGRAPH':
                    break
                one_intro_list.append(line)
            one_intro = " ".join(one_intro_list)
        except:
            print(full_path,'no introduction')
            one_intro = ""
        introduction.append(one_intro)
        
        # 处理conclusion
        full_path = path+'FULLTEXT'
        data = open(full_path)
        lines = data.readlines()
        lines = [i.strip('\n') for i in lines]
        lines = [i for i in lines if i != '']
        try: 
            lines = lines[lines.index('Conclusions')+2:]
            one_conclu_list = []
            for line in lines:
                if line == 'PARAGRAPH':
                    break
                one_conclu_list.append(line)
            one_conclu = " ".join(one_conclu_list)
        except:
            print(full_path,'no conclusion')
            one_conclu = ""
        conclusion.append(one_conclu)
        
        # 处理summary
        laysumm_path = path+'LAYSUMM'
        data = open(laysumm_path)
        lines = data.readlines()
        lines = [i.strip('\n') for i in lines]
        lines = [i for i in lines if i != '']
        lines = lines[lines.index('PARAGRAPH')+1:]
        one_summ = " ".join(lines)
        summary.append(one_summ)
    return abstract, introduction, summary, conclusion
```

```python
abstract_1, introduction_1, summary_1, conclusion_1 = process_data(batch_1_list)
abstract_2, introduction_2, summary_2, conclusion_2 = process_data(batch_2_list)
```

```python
abstract = abstract_1 + abstract_2
introduction = introduction_1 + introduction_2
conclusion = conclusion_1 + conclusion_2
summary = summary_1 + summary_2
def cal_mean_len(l):
    length = [len(i.split()) for i in l if i!='']
    import numpy as np
    print(np.mean(length))
cal_mean_len(abstract)
cal_mean_len(introduction)
cal_mean_len(conclusion)
cal_mean_len(summary)
```

```python
index=list(range(0,len(abstract)))
chosen_idex = random.sample(index, int(len(abstract)*0.1))

train_source = open('./processed_data/train.source','w')
test_source = open('./processed_data/valid.source','w')
train_target = open('./processed_data/train.target','w')
test_target = open('./processed_data/valid.target','w')

for i in range(len(abstract)):
    if i in chosen_idex:
        # 选择为测试集
        if abstract[i] == "":
            continue
            test_source.write(introduction[i]+'\n')
        else:
            test_source.write(abstract[i]+''+introduction[i] +'\n')
        test_target.write(summary[i]+'\n')
        
    else:
        # 选择为训练集
        if abstract[i] == "":
            continue
            train_source.write(introduction[i]+'\n')
        else:
            train_source.write(abstract[i]+''+introduction[i] +'\n')
        train_target.write(summary[i]+'\n')
        

train_source.close()
test_source.close()
train_target.close()
test_target.close()
```
